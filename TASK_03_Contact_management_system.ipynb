```python
%%writefile app.py
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Contact Management System", page_icon="ðŸ“‡")
st.title("ðŸ“‡ Contact Management System")

# Initialize contacts
if "contacts" not in st.session_state:
    st.session_state.contacts = []

# ---------------- ADD CONTACT ----------------
st.subheader("âž• Add New Contact")

with st.form("add_form", clear_on_submit=True):
    name = st.text_input("Name")
    phone = st.text_input("Phone Number")
    email = st.text_input("Email Address")
    add_btn = st.form_submit_button("Add Contact")

    if add_btn:
        if name and phone and email:
            st.session_state.contacts.append({
                "name": name,
                "phone": phone,
                "email": email
            })
            st.success("âœ… Contact added successfully")
        else:
            st.warning("âš ï¸ Fill all fields")

# ---------------- VIEW / EDIT / DELETE ----------------
st.subheader("ðŸ“‹ View Contacts")

if st.button("Show Contacts"):
    if not st.session_state.contacts:
        st.info("No contacts available")
    else:
        for i, c in enumerate(st.session_state.contacts):
            with st.expander(f"ðŸ‘¤ {c['name']}"):
                new_name = st.text_input("Edit Name", c["name"], key=f"name{i}")
                new_phone = st.text_input("Edit Phone", c["phone"], key=f"phone{i}")
                new_email = st.text_input("Edit Email", c["email"], key=f"email{i}")

                col1, col2 = st.columns(2)

                with col1:
                    if st.button("âœï¸ Update", key=f"update{i}"):
                        st.session_state.contacts[i] = {
                            "name": new_name,
                            "phone": new_phone,
                            "email": new_email
                        }
                        st.success("Updated successfully")

                with col2:
                    if st.button("ðŸ—‘ Delete", key=f"delete{i}"):
                        st.session_state.contacts.pop(i)
                        st.warning("Contact deleted")
                        st.experimental_rerun()

# ---------------- SAVE TO FILE ----------------
st.subheader("ðŸ’¾ Save Contacts")

if st.button("Save to File"):
    if st.session_state.contacts:
        df = pd.DataFrame(st.session_state.contacts)
        df.to_csv("contacts.csv", index=False)
        st.success("ðŸ“ Contacts saved as contacts.csv")
    else:
        st.info("No contacts to save")
```
